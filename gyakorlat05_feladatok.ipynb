````sql
SELECT Rendeles.SZALL_DATUM, Rendeles.SZALL_MOD, SUM(Rendeles_tetel.MENNYISEG)
FROM Rendeles JOIN Rendeles_tetel ON Rendeles.SORSZAM=Rendeles_tetel.SORSZAM
JOIN Termek ON Rendeles_tetel.TERMEKKOD=Termek.TERMEKKOD
WHERE MEGYS='db'
GROUP BY ROLLUP(Rendeles.SZALL_DATUM, Rendeles.SZALL_MOD)
````
